```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.__version__
from mpl_toolkits.mplot3d import Axes3D

from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
def makePlot(simHits, simClusters, event):
    fig = plt.figure(figsize=[10, 10])
    ax = fig.add_subplot(111, projection='3d')
    colormap = {-1 : 'grey', 111 : "red", 211 : 'blue', 11 : 'green', 13 : 'orange', 22 : "lightblue", 
                    2112 : "pink", 2212 : "pink"}
    nsc = len(simClusters.SimCluster_impactPointX[event])
    print("Number of SimClusters", nsc)
    for i in range(-1, nsc):
        if i == -1:
            pdgId = -1
        else:
            scIdx = simHits.SimHitHGCEE_SimClusterIdx[event,i]
            pdgId = abs(simClusters.SimCluster_pdgId[event,scIdx])
        if not len(simHits[simHits.SimHitHGCEE_SimClusterIdx == i].SimHitHGCEE_x):
            continue
        x = simHits[simHits.SimHitHGCEE_SimClusterIdx == i].SimHitHGCEE_x[event]
        y = simHits[simHits.SimHitHGCEE_SimClusterIdx == i].SimHitHGCEE_y[event]
        z = simHits[simHits.SimHitHGCEE_SimClusterIdx == i].SimHitHGCEE_z[event]
        color = colormap[pdgId]
        ax.scatter(z,x,y, marker='o', zorder=i, c=color if i != -1 else 'gray',alpha=0.25)
    scX = simClusters.SimCluster_impactPointX[event]
    scY = simClusters.SimCluster_impactPointY[event]
    scZ = simClusters.SimCluster_impactPointZ[event]
    ax.scatter(scZ,scX,scY, marker='x', c='black', s=500, zorder=nsc+2)
    ax.set_xlim3d(-500, 500)
    ax.set_ylim3d(-150,150)
    ax.set_zlim3d(-150,150)
    ax.set_xlabel('z pos')
    ax.set_ylabel('x pos')
    ax.set_zlabel('y pos')

    return fig
```

```python
fig = makePlot(simHits, simClusters, 4)
```

```python
fig = makePlot(simHitsDefault, simClustersDefault, 4)
```
